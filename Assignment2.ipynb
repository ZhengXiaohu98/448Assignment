```python
#Xiaohu Zheng
# Solving a leetcode problem
# You are given an integer array nums.
# The range of a subarray of nums is the difference between the largest and smallest element in the subarray.

# Return the sum of all subarray ranges of nums.

# Example:
# Input: nums = [1,2,3]
# Output: 4
# Explanation: The 6 subarrays of nums are the following:
# [1], range = largest - smallest = 1 - 1 = 0 
# [2], range = 2 - 2 = 0
# [3], range = 3 - 3 = 0
# [1,2], range = 2 - 1 = 1
# [2,3], range = 3 - 2 = 1
# [1,2,3], range = 3 - 1 = 2
# So the sum of all ranges is 0 + 0 + 0 + 1 + 1 + 2 = 4.

def subArrayRanges(nums: list[int]) -> int:
        n = len(nums)
        minL, minR, maxL, maxR = [0]*n, [0]*n, [0]*n, [0]*n
        minStk, maxStk = [], []
        for i, x in enumerate(nums):
            while minStk and x < nums[minStk[-1]]:
                minStk.pop()
            minL[i] = minStk[-1] if minStk else -1
            minStk.append(i)

            while maxStk and x > nums[maxStk[-1]]:
                maxStk.pop()
            maxL[i] = maxStk[-1] if maxStk else -1
            maxStk.append(i)

        minStk, maxStk = [], []
        for i in range(n-1, -1, -1):
            while minStk and nums[i] <= nums[minStk[-1]]:
                minStk.pop()
            minR[i] = minStk[-1] if minStk else n
            minStk.append(i)

            while maxStk and nums[i] >= nums[maxStk[-1]]:
                maxStk.pop()
            maxR[i] = maxStk[-1] if maxStk else n
            maxStk.append(i)

        maxSum, minSum = 0, 0
        for i in range(n):
            minSum += nums[i] * (i-minL[i]) * (minR[i]-i)
            maxSum += nums[i] * (i-maxL[i]) * (maxR[i]-i)
        return maxSum - minSum
    
nums = [1,2,3]
print("The sum is: ", subArrayRanges(nums))
```

##### Explaination

- This is an O(n) time complexity solution
- We don't really care what the subset is, we just need to know how many times the number are used as minimum, and how many times are used as maximum.
- The idea behind is that: For each number, we need to find the minL, minR (the smaller number on the left, the smaller number on the right), we need to do the same for maxL, maxR.
- The number of times used as minimum is calculated as `(i - minL[i]) * (minR[i] - i)`, this is very easy to demonstate, for example we have an array `[0, 7, 5, 6, 4]` and we are trying to find how many times `5` is used as minimum.
- `5` is at index of 2, and its minL is `0` which is at index 0, and minR is `4` which is at index 4. So it is used `(2-0) * (4-2) = 2*2 = 4` times.
- Care point: we need to becareful that for right we find the number that is strickly smaller and larger, so for left we stop if the numbers are equal. We cannot have both at the same time!!!
